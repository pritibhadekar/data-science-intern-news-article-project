```python
import feedparser

def parse_rss_feeds(feed_urls):
    all_articles = []

    for feed_url in feed_urls:
        feed = feedparser.parse(feed_url)

        for entry in feed.entries:
            article = {
                'title': entry.title,
                #'content': entry.summary,
                'pub_date': entry.get('published_parsed', ''),
                'source_url': entry.link,
                'feed_url': feed_url
            }

            # Check for duplicate articles
            if article not in all_articles:
                all_articles.append(article)

    return all_articles

# Example list of RSS feed URLs
rss_feeds = [
    'http://rss.cnn.com/rss/cnn_topstories.rss',
    'http://qz.com/feed',
    'http://feeds.foxnews.com/foxnews/politics',
    'http://feeds.reuters.com/reuters/businessNews',
    'http://feeds.feedburner.com/NewshourWorld',
    'https://feeds.bbci.co.uk/news/world/asia/india/rss.xml'
    # Add more feed URLs as needed
]

# Run the script
result = parse_rss_feeds(rss_feeds)

# Print the extracted information
for idx, article in enumerate(result, start=1):
    print(f"Article {idx} - {article['title']}")
    print(f"Publication Date: {article['pub_date']}")
    print(f"Source URL: {article['source_url']}")
    #print(f"Content: {article['content']}")
    print(f"Feed URL: {article['feed_url']}")
    print("-" * 50)
```

```python
# Install SQLAlchemy if not already installed
!pip install sqlalchemy

from sqlalchemy import create_engine, Column, String, Text, DateTime, Integer, exc
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy.orm import sessionmaker
from datetime import datetime

# Define the database connection string
db_connection_string = 'mysql://root:123456@127.0.0.1:3307/News_Articles'  # SQLite example, replace with your database connection string

# Create the engine and connect to the database
engine = create_engine(db_connection_string, echo=True)

# Define the Article model
Base = declarative_base()

class Article(Base):
    __tablename__ = 'articles'
    ArticleID = Column(Integer, primary_key=True, autoincrement=True)
    Title = Column(String, unique=True)  # Make the title unique to prevent duplicates
    Content = Column(Text)
    PublicationDate = Column(DateTime)
    Source = Column(String)

# Create the table
Base.metadata.create_all(engine)

# Function to store new articles without duplicates
def store_article(title, content, publication_date, source):
    Session = sessionmaker(bind=engine)
    session = Session()

    try:
        # Attempt to create a new article
        new_article = Article(
            Title=title,
            Content=content,
            PublicationDate=publication_date,
            Source=source
        )
        session.add(new_article)
        session.commit()
        print("Article stored successfully.")
    except exc.IntegrityError as e:
        # IntegrityError occurs when there's a duplicate title
        session.rollback()
        print(f"Duplicate article. Not storing. Error: {e}")
    finally:
        session.close()

# Example usage
article_title = "Breaking News"
article_content = "This is a breaking news article."
article_publication_date = datetime.now()
article_source = "CNN"

store_article(article_title, article_content, article_publication_date, article_source)
```
